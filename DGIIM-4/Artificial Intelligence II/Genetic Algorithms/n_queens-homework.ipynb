To solve the $N-$ queens problem, we will use a genetic algorithm. The algorithm will be based on the following steps:
1. Generate a random population of $N-$ queens.
2. Evaluate the fitness of each individual.
3. Select the best individuals.
4. Apply crossover and mutation to the selected individuals.
5. Repeat steps 2-4 until a solution is found.
6. If no solution is found, repeat steps 1-5.

To achieve this, we will begin importing the necessary libraries:

```python
import pygad
import numpy as np
```

We will use the *pygad* library which is a Python library for genetic algorithms. Now, we will define the parameters of the genetic algorithm:

```python
n = 8 # The number of queens.
population_size = 100 # The number of solutions in the population.
fmax = sum([i for i in range(n)])-1 # The maximum fitness value.
reach = f"reach_{fmax}"
```

Next, we will define the fitness function. The fitness function will be the number of non-repeated pairs of queens that are not attacking each other. The higher the fitness value, the better the solution. The fitness function will be applied to each individual of the population.

```python
# Fitness function for the genetic algorithm.
def fitness_function(chromosome, chromosome_idx=None):
    # Non-attacking pairs of queens have a fitness value of 1.
    # Attacking pairs of queens have a fitness value of 0.
    fitness = 0
    for i in range(n):
        for j in range(i+1,n):
            if chromosome[i] != chromosome[j]:
                dx = abs(i - j)
                dy = abs(chromosome[i] - chromosome[j])
                if dx != dy:
                    fitness += 1
    return fitness
```

Now, we will construct an object of the class GA. This class contains all the necessary functions to run the genetic algorithm. It accepts many parameters, but we will only use the following ones:
* *num_generations*: The number of generations.
* *num_parents_mating*: The number of parents that will be used to produce offspring.
* *fitness_func* : The fitness function.
* *parent_selection_type*: The type of parent selection. We will use *roulette_wheel_selection*.
* *crossover_type*: The type of crossover. We will use *single_point_crossover*.
* *sol_per_pop*: The number of individuals in the population.
* *num_genes*: The number of genes in each individual. In this case, the number of genes is the number of queens.
* *gene_type*: The type of genes. We will use *int*.
* *init_range_low*: The lower bound of the initial population.
* *init_range_high*: The upper bound of the initial population.
* *mutation_probability*: The probability of mutation of a gene.
* *mutation_percent_genes*: The percentage of genes that will be mutated.
* *stop_criteria*: The criteria to stop the algorithm. We will use *fmax*, which is the max number of not attacking pairs of queens.

We could have also set a concrete initial population, but we will let the algorithm generate it randomly.

```python
ga_instance = pygad.GA(num_generations=500, 
                       num_parents_mating=20,
                       fitness_func=fitness_function,
                       parent_selection_type="rws",
                       crossover_type="single_point",
                       sol_per_pop=population_size,
                       num_genes=n,
                       gene_type = int,
                       init_range_low=1,
                       init_range_high=n,
                       mutation_probability=0.25,
                       mutation_percent_genes=100,
                       stop_criteria=reach
                       )
```

Once created the object, we will run the genetic algorithm and plot the results:

```python
ga_instance.run()
ga_instance.plot_fitness()
```

Finally, this is the solution found by the genetic algorithm:

```python
ga_instance.best_solution()
```
